```python
# List of numbers
num = [1, 2, 3, 4, 5, 3, 4, 52, 5, 7, 83, 46, 3, 4, 6, 7, 9, 7, 2, 3, 4, 6, 689]

# List of strings
str_list = [
    'a','s','f','g','h','a','w','e','r','f',
    'as','cz','x','c','vsd','wa','a','sf','zxxf','a',
    'as','fe','zd','e','f','w','af','h','j','k','o','u','g','f','d','f','n','u','ui','j','h'
]

def uniq(numbers, strings):
    # Unique numbers
    unique_numbers = list(set(numbers))
    
    # Unique strings, sorted
    seen = set()
    unique_strings = []
    for s in sorted(strings):  # first sort
        if s not in seen:
            unique_strings.append(s)
            seen.add(s)
    
    return unique_numbers, unique_strings

# Usage
nums, strs = uniq(num, str_list)
print(nums)
print(strs)
```

```python
def is_palindrome(word):
    stack =[]

    for char in word:
        stack.append(char)

    for char in word:
        if char!= stack.pop():
            return False
        return True


    
```

```python
from itertools import permutations

word = "listen"

# itertools.permutations returns tuples of letters
all_perms = permutations(word)

# Convert each tuple to string and store in a set (to get unique anagrams)
anagrams = set(''.join(p) for p in all_perms)

# Remove the original word if you want
anagrams.discard(word)

print(anagrams)
```

```python
from matplotlib import animation, rc
import matplotlib.pyplot as plt
import random
import seaborn as sns

# --- Update function for FuncAnimation ---
def update(frame_number, rolls, faces, frequencies):
    # Roll the die "rolls" times
    for i in range(rolls):
        frequencies[random.randrange(1, 7) - 1] += 1

    plt.cla()  # clear the axes
    axes = plt.gca()  # get current axes
    sns.barplot(x=faces, y=frequencies, palette='bright', ax=axes)

    axes.set_title(f'Die frequencies for {sum(frequencies):,} Rolls')
    axes.set(xlabel='Die Value', ylabel='Frequencies')
    axes.set_ylim(top=max(frequencies) * 1.15)

    # Add count + percentage above each bar
    for bar, frequency in zip(axes.patches, frequencies):
        text_x = bar.get_x() + bar.get_width() / 2.0
        text_y = bar.get_height()
        text = f'{frequency:,}\n{frequency / sum(frequencies):.3%}'
        axes.text(text_x, text_y, text, ha='center', va='bottom')

# --- Parameters ---
number_of_frames = 10    # number of animation steps
rolls_per_frame = 100    # rolls added per frame
sns.set_style('whitegrid')

# --- Initialize figure and data ---
figure = plt.figure('Rolling a Six-Sided Die')
values = list(range(1, 7))
frequencies = [0] * 6

# --- Animation setup ---
rc('animation', html='jshtml')  # render animation as HTML5 in Jupyter

die_animation = animation.FuncAnimation(
    figure,
    update,
    repeat=False,
    frames=number_of_frames,
    interval=500,  # ms per frame
    fargs=(rolls_per_frame, values, frequencies)
)

die_animation  # <-- in Jupyter, display the animation inline
```
