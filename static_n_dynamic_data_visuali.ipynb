```python
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns

# simulate dice rolls
rolls = [random.randrange(1, 7) for i in range(600)]
values, frequencies = np.unique(rolls, return_counts=True)

# title
title = f'Rolling Six-Sided Die {len(rolls)} Times'

# seaborn style
sns.set_style('whitegrid')

# barplot
axes = sns.barplot(x=values, y=frequencies, palette=sns.color_palette("bright", len(values)))

# labels
axes.set_title(title)
# axes.set_xlabel("Die Value")
# axes.set_ylabel("Frequencies")
axes.set(xlabel = 'Die Value', ylabel = 'frequencies')

# show plot
plt.show()
```

```python
# List of numbers
num = [1, 2, 3, 4, 5, 3, 4, 52, 5, 7, 83, 46, 3, 4, 6, 7, 9, 7, 2, 3, 4, 6, 689]

# List of strings
str_list = [
    'a','s','f','g','h','a','w','e','r','f',
    'as','cz','x','c','vsd','wa','a','sf','zxxf','a',
    'as','fe','zd','e','f','w','af','h','j','k','o','u','g','f','d','f','n','u','ui','j','h'
]

def uniq(numbers, strings):
    # Unique numbers
    unique_numbers = list(set(numbers))
    
    # Unique strings, sorted
    seen = set()
    unique_strings = []
    for s in sorted(strings):  # first sort
        if s not in seen:
            unique_strings.append(s)
            seen.add(s)
    
    return unique_numbers, unique_strings

# Usage
nums, strs = uniq(num, str_list)
print(nums)
print(strs)
```

```python
def is_palindrome(word):
    stack =[]

    for char in word:
        stack.append(char)

    for char in word:
        if char!= stack.pop():
            return False
        return True


    
```

```python
lst = ['r', 'a', 'd', 'a', 'r']
lst2 = ['r', 'a', 'd']
print(is_palindrome(lst))
print(is_palindrome(lst2))
```

```python
from itertools import permutations

word = "listen"

# itertools.permutations returns tuples of letters
all_perms = permutations(word)

# Convert each tuple to string and store in a set (to get unique anagrams)
anagrams = set(''.join(p) for p in all_perms)

# Remove the original word if you want
anagrams.discard(word)

print(anagrams)
```

```python
import string

def is_pangram(text):
    text = text.lower()
    return all(letter in text for letter in string.ascii_lowercase)

print(is_pangram("The quick brown fox jumps over the lazy dog"))  # True
```
